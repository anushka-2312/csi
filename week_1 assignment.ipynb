```python
#1 Loading dataset
#importing the required libraries 
import pandas as pd
df = pd.read_csv("C:\\Users\\anush\\Downloads\\shopping\\Combined_dataset.csv")
#verfiying the data by first few rows
df.head()
```

```python
#2. Exploring data 
#printing first five rows by head() method
print("FIRST FIVE ROWS")
print(df.head())
#printing the last five rows by tail() method 
print("LAST FIVE ROWS")
print(df.tail())
```

```python
#checking for the rows and column by shape
print(f"\nDataset Shape: {df.shape}")

#the columns in dataset and the data type
print(f"\n Columns: {df.columns.tolist}")
print(f"\n Data types")
df.info()
```

```python
#3. Handling the Missing Values

#Identifying the missing values
print("Missing values per column before:")
print(df.isnull().sum())

#Managing by either drop or fill missing values 
if 'discount' in df.columns:
    df['discount'] = df['discount'].fillna(0)

df['seller_name'] = df['seller_name'].fillna('Unknown Seller')
df['seller_information'] = df['seller_information'].fillna('No Seller Info Available')
df['variations'] = df['variations'].fillna('No variations')
df['what_customers_said'] = df['what_customers_said'].fillna('No customer feedback available')
df.drop(columns=['videos'], inplace=True)

print(f"\nAfter handling he missing values")
print(df.isnull().sum())
```

```python
#4.perform basic operation(filter rows, select columns)
#for the highly rated products
high_rated_products = df[df['rating'] > 4.5]

print(f"Number of products with rating above 4.5: {high_rated_products.shape[0]}")

#some specific columns
product_pricing = df[['title', 'initial_price', 'rating']]
print(product_pricing.head())
```

```python
#5. Remove Duplicates
duplicate_count = df.duplicated().sum()
print(f"Total number of duplicate rows: {duplicate_count}")

#if duplicates present then remove them 
df.drop_duplicates(inplace=True)

#after removing duplicates checking the 
print(f"Dataset shape after removing duplicates: {df.shape}")
```

```python
#6.Create a derived column (total_amount = price * quantity)
df['quantity'] = 2
df['total_amount'] = df['initial_price'] * df['quantity']
print(df[['title', 'initial_price', 'quantity', 'total_amount']].head())
```

```python
# 7.Save the cleaned dataset as a new CSV file. Output: Jupyter Notebook (.ipynb) + cleaned CSV + brief summary

df.to_csv('combined_dataset_cleaned.csv', index=False)

print("Successfully cleaned the dataset and saved")
```
